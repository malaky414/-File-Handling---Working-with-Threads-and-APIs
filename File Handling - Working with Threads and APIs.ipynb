```python
import threading
import requests
def download_file(url, filename):
    print(f"ğŸ“¥ Starting download: {filename}")
    response = requests.get(url)
    with open(filename, 'wb') as f:
        f.write(response.content)
    print(f"âœ… Finished download: {filename}")
urls = [
    ("https://www.google.com/url?sa=i&url=https%3A%2F%2Fwww.tasteofhome.com%2Frecipes%2Fhomemade-pizza%2F&psig=AOvVaw3iWvuATy-KUx0GanVNGFDR&ust=1751265230795000&source=images&cd=vfe&opi=89978449&ved=0CBQQjRxqFwoTCLD4_8GBlo4DFQAAAAAdAAAAABAE", "image1.png"),
    ("https://www.google.com/url?sa=i&url=https%3A%2F%2Fcookidoo.international%2Frecipes%2Frecipe%2Fen%2Fr754564&psig=AOvVaw3iWvuATy-KUx0GanVNGFDR&ust=1751265230795000&source=images&cd=vfe&opi=89978449&ved=0CBQQjRxqFwoTCLD4_8GBlo4DFQAAAAAdAAAAABAL", "image2.png"),
    ("https://www.google.com/url?sa=i&url=https%3A%2F%2Feu.gozney.com%2Fblogs%2Frecipes%2Fitalian-sausage-and-hot-honey-pizza-recipe&psig=AOvVaw3iWvuATy-KUx0GanVNGFDR&ust=1751265230795000&source=images&cd=vfe&opi=89978449&ved=0CBQQjRxqFwoTCLD4_8GBlo4DFQAAAAAdAAAAABAW", "image3.png"),
]
threads = []

for url, filename in urls:
    thread = threading.Thread(target=download_file, args=(url, filename))
    thread.start()
    threads.append(thread)

for thread in threads:
    thread.join()

print("ğŸ‰ All downloads completed!")
```

```python
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)
if response.status_code == 200:
    posts = response.json()
    for post in posts:
        print("ğŸ“Œ", post["title"])
else:
    print("âŒ Failed to fetch posts")
```

```python
def fetch_data(url, label):
    print(f"ğŸŒ Fetching {label}...")
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        print(f"âœ… {label} fetched: {len(data)} items")
    else:
        print(f"âŒ Failed to fetch {label}")
apis = [
    ("https://jsonplaceholder.typicode.com/posts", "Posts"),
    ("https://jsonplaceholder.typicode.com/comments", "Comments"),
    ("https://jsonplaceholder.typicode.com/users", "Users"),
]
threads = []

for url, label in apis:
    thread = threading.Thread(target=fetch_data, args=(url, label))
    thread.start()
    threads.append(thread)
for thread in threads:
    thread.join()

print("ğŸ‰ All API data fetched!")
```
